```python
print(df.isnull().sum())
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown Cast', inplace=True)
df['country'].fillna('Unknown Country', inplace=True)
df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['type'] = df['type'].str.strip() 
```

```python
df['type'] = df['type'].str.title() 
df['rating'] = df['rating'].str.upper()  
```

```python
print(df['type'].unique())  # Should show consistent values like ['Movie', 'TV Show']
print(df['rating'].unique())  # Should show consistent ratings like ['TV-MA', 'PG', etc.]
print(df['country'].unique())  # Should show properly capitalized country names
print(df['listed_in'].unique())  # Should show consistent categories
```

```python
print(df['director'].str.strip().eq(df['director']).all())  # Should print True
print(df['cast'].str.strip().eq(df['cast']).all())  # Should print True
print(df['description'].str.strip().eq(df['description']).all())  # Should print True
```

```python
# Extract numeric part of the 'duration' column (e.g., '90 min' -> 90)
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(float)
```

```python
print(df['duration_num'].head())  # View the first few rows of 'duration_num' to verify
```

```python
print(df['duration_num'].min(), df['duration_num'].max())  # Check the min and max of 'duration_num'
```

```python
# Check if there are missing values in 'duration_num'
print(df['duration_num'].isnull().sum())  # Count missing values in 'duration_num'

# Optionally, fill missing values with a default value (e.g., 0)
df['duration_num'].fillna(0, inplace=True)  # Replace NaNs with 0

# Or drop rows with missing values
# df.dropna(subset=['duration_num'], inplace=True)  # Drop rows with NaN in 'duration_num'
```
